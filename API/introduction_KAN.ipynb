# Знакомство с KAN

### Теорема Колмогорова-Арнольда о представлении функции

Теорема Колмогорова-Арнольда о представлении сложной функции утверждает, что если $f$ - многомерная непрерывная функция ограниченная на некоторой области, то она может быть записана как конечная композиция непрерывных одномерных функций и бинарной операции сложения. Т.е. для гладкой функции $f: [0, 1]^n \to \mathbb{R}$,

$$f(x) = f(x_1, x_2, ..., x_n) = \sum_{q=1}^{2n+1}\Phi_q(\sum_{p=1}^{n}\phi_{q, p}(x_p))$$

где $\phi_{q,p}:[0,1]\to\mathbb{R}$ и $\Phi_q:\mathbb{R}\to\mathbb{R}$.


### Сети Колмогорова-Арнольда (KAN)

Представление Колмогорова-Арнольда может быть записано в матричной форме

$$f({\bf x}) = {\bf \Phi}_{\rm out}\circ{\bf \Phi}_{\rm in}\circ\bf{x}$$

где

$${\bf \Phi}_{\rm in}= \begin{pmatrix} \phi_{1,1}(\cdot) & \cdots & \phi_{1,n}(\cdot) \\ \vdots & & \vdots \\ \phi_{2n+1,1}(\cdot) & \cdots & \phi_{2n+1,n}(\cdot) \end{pmatrix},\quad {\bf \Phi}_{\rm out}=\begin{pmatrix} \Phi_1(\cdot) & \cdots & \Phi_{2n+1}(\cdot)\end{pmatrix}$$
